```python
E_pos = []
grouped = sgldrg[['DRUG_CONCEPT_NAME','EVENT_CONCEPT_NAME','EVENT_TYPE']]
# grouped = grouped[grouped.DRUG_CONCEPT_NAME == 363] #363
grouped['drugid'] = 434

grouped = grouped.groupby('DRUG_CONCEPT_NAME')
for name, group in grouped:
    ls = []
    dim1_df = group[['drugid','DRUG_CONCEPT_NAME']].drop_duplicates()
    dim2_df = group[['EVENT_TYPE','EVENT_CONCEPT_NAME']].drop_duplicates()
    p = np.append(dim1_df.values,dim2_df.values,0)
    
    if (len(p)>512):
        p = p[:512]
        new_list = p
    else:
        counter = 512 - len(p)
        while (counter > 0):
            p = np.append(p,[0,0])
            counter -= 1

        new_list = [list(p[i:i+2]) for i in range(0, len(p), 2)]
    E_pos.append(np.array(new_list))
```

```python
ddidrg_lgt = ddidrg_lgt[ddidrg_lgt.EVENT_CONCEPT_NAME.notna()]
ddidrg_lgt
```

```python
# E_pos = []
grouped = ddidrg_lgt[['DRUG_1_CONCEPT_NAME','DRUG_2_CONCEPT_NAME','EVENT_CONCEPT_NAME']]
grouped['drugid'] = 434
grouped['adrid'] = 2381

grouped = grouped.groupby(['DRUG_1_CONCEPT_NAME','DRUG_2_CONCEPT_NAME'])
for name, group in grouped:
    ls = []
    dim1_df = group[['drugid','DRUG_1_CONCEPT_NAME']].drop_duplicates()
    dim2_df = group[['drugid','DRUG_2_CONCEPT_NAME']].drop_duplicates()
    dim3_df = group[['adrid','EVENT_CONCEPT_NAME']].drop_duplicates()
    p = np.append(dim1_df.values,dim2_df.values,0)
    p = np.append(p,dim3_df.values,0)
    
    if (len(p)>512):
        p = p[:512]
        new_list = p
    else:
        counter = 512 - len(p)
        while (counter > 0):
            p = np.append(p,[0,0])
            counter -= 1

        new_list = [list(p[i:i+2]) for i in range(0, len(p), 2)]
    E_pos.append(np.array(new_list))
```

```python
E_neg = []
adr_id = sgldrg[sgldrg.EVENT_TYPE == 2381].EVENT_CONCEPT_NAME.to_numpy()
ind_id = sgldrg[sgldrg.EVENT_TYPE == 1557].EVENT_CONCEPT_NAME.to_numpy()
for ix in range(len(E_pos_)):
    data = E_pos_[ix]
    ##random get values from numpy array with fixed length
    M = data[data[:, 0] == 2381, 1]
    data[data[:, 0] == 2381, 1] = np.random.choice(adr_id, len(M), replace=False)
    # change indication
    N = data[data[:, 0] == 1557, 1]
    data[data[:, 0] == 1557, 1] = np.random.choice(ind_id, len(N), replace=False)

    E_neg.append(data)
```

```python
#转换为邻接矩阵用pivot（）
vals = np.unique(sgldrg[['DRUG_CONCEPT_NAME', 'EVENT_CONCEPT_NAME']])  # 同时取出两列,作为节点
df = pd.DataFrame(0, index=vals, columns=vals)
f = df.index.get_indexer
df.values[f(sgldrg.DRUG_CONCEPT_NAME), f(sgldrg.EVENT_CONCEPT_NAME)] = 1
print(df)
```

```python
G = nx.from_numpy_matrix(np.array(df))
A_lapl = nx.normalized_laplacian_matrix(G).toarray()
print("\n Laplacian normalization")
print(A_lapl)
```

```python
path = "gadj.pkl"
obj = A_lapl
with open(path, 'wb') as f:
    pickle.dump(obj,f)
```
